```python
#Bubble sort
def bubblesort(arr):
  # Looping from size of array from last index[-1] to index [0]
  for n in range(len(arr)-1, 0, -1):
    for i in range(n):
      print('For i=',n,'  j=',i,arr)
      if arr[i] > arr[i + 1]:
        # swapping data if the element is less than next element in the array
        arr[i], arr[i + 1] = arr[i + 1], arr[i]


def partition(arr, low, high):
    i = (low-1)         # index of smaller element
    pivot = arr[high]     # pivot
 
    for j in range(low, high):
 
        # If current element is smaller than or
        # equal to pivot
        if arr[j] <= pivot:
 
            # increment index of smaller element
            i = i+1
            arr[i], arr[j] = arr[j], arr[i]
 
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return (i+1)
 

 
# Function to do Quick sort
 

def quickSort(arr, low, high):
    global iteration
    iteration+=1
    print('Iteration :',iteration,' ',arr)
    # print(arr)
    if len(arr) == 1:
        return arr
    if low < high:
 
        # pi is partitioning index, arr[p] is now
        # at right place
        pi = partition(arr, low, high)
        # Separately sort arr before
        # partition and after partition
        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)
 

       
arr = [39,12,18,85,72,10,2,18,6,4]
  
print("Unsorted list is, ",arr) 
print("------------Bubble Sort--------------")
bubblesort(arr)
print("Sorted Array is using bubble Sort, ")
print(arr)



# Driver code to test above
iteration=0 
arr = [3, 44, 6, 1, 88, 98,77, 34, 5, 4]
print("Unsorted list is, ",arr)
print("------------Quick Sort--------------") 
n = len(arr)
quickSort(arr, 0, n-1)
print("Sorted array is using Quick sort:")
print(arr)
```
